```python
df = df.rename(columns={
    'nameOrig': 'User_ID',
    'nameDest': 'Recipient_ID',
    'isFraud': 'Status_Code',         
    'type': 'Transaction_Type',
    'amount': 'Transaction_Amount',
    'oldbalanceOrg': 'Sender_Balance_Before',
    'newbalanceOrg': 'Sender_Balance_After',
    'oldbalanceDest': 'Receiver_Balance_Before',
    'newbalanceDest': 'Receiver_Balance_After'
}).copy()
```

```python
df.drop('isFlaggedFraud', axis=1, inplace=True)
```

```python
df
```

```python
# Map the Status_Code (0: SUCCESS, 1: FAILED)
# 0 (Zero): Represents the normal, baseline, or desired state. In a fraud context, this means Not Fraudulent or Legitimate.
# 1 (One): Represents the target, abnormal, or undesired event you are trying to predict or measure. In this context, this means Fraudulent

df['Transaction_Status'] = df['Status_Code'].map({0: 'SUCCESS', 1: 'FAILED'})
```

```python
# --- ESSENTIAL FILTERING CODE ---
# 1. Count how many rows have a $0.00 amount
zero_amount_count = len(df[df['Transaction_Amount'] == 0.00])
```

```python
df = df[df['Transaction_Amount'] > 0.00].copy()
```

```python
df
```

```python
df[df['Transaction_Amount'] <= 0.00].shape[0] #  df contained zero rows where the 'Transaction_Amount' was less than or equal to $$$0.00.
```

```python
# 1. Calculate Transaction Success Rate (TSR)

# Assuming the 'Transaction_Status' column exists and 'SUCCESS' is the status for a successful transaction.
total_transactions = len(df)
successful_transactions = len(df[df['Transaction_Status'] == 'SUCCESS'])
# Calculate the Success Rate as a percentage
success_rate = (successful_transactions / total_transactions) * 100
```

```python
print(f"\n--- KPI 1: Transaction Success Rate (TSR) ---")
print(f"Total Final Transactions: {total_transactions}")
print(f"Successful Transactions: {successful_transactions}")
print(f"Transaction Success Rate: {success_rate:.2f}%") # just to check 
```

```python
# 2. Calculate Average Transaction Value (ATV)

# Filter for only successful transactions.
# We use .copy() to ensure we are working with an independent DataFrame.
df_success = df[df['Transaction_Status'] == 'SUCCESS'].copy()

# Calculate the average amount of successful transactions
avg_transaction_value = df_success['Transaction_Amount'].mean() # just for check
```

```python
print(f"\n--- KPI 2: Average Transaction Value (ATV) ---")
print(f"Average Transaction Value (ATV): ${avg_transaction_value:,.2f}") # just for check
```

```python
# 3. Basic User Segmentation
user_segmentation = df.groupby('User_ID').agg(
    transaction_count=('Transaction_Amount', 'count'),
    total_amount_sent=('Transaction_Amount', 'sum'),
    average_amount_sent=('Transaction_Amount', 'mean')
).reset_index()

# Merge the segmentation features back into the main DataFrame
df = df.merge(user_segmentation, on='User_ID', how='left')
```

```python
# Print columns to confirm success
print("Columns after successful merge:")
print(df.columns.tolist())
```

```python
# --- PHASE 3: FINAL MODEL PREPARATION ---

# 1. One-Hot Encode the 'Transaction_Type' column
# This converts categories (PAYMENT, TRANSFER, etc.) into numerical features
df = pd.get_dummies(df, columns=['Transaction_Type'], drop_first=True)

# 2. Drop High-Cardinality and Unnecessary Columns
# We drop the ID columns (User_ID, Recipient_ID) and the redundant text status column.
df = df.drop(columns=['User_ID', 'Recipient_ID', 'Transaction_Status'])

print("✅ Categorical Encoding and Cleanup Complete.")
print("The data is now ready for scaling and model training.")
print("Remaining features:", df.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# --- PHASE 3: MODEL BUILDING ---

# 1. Define Features (X) and Target (y)
# 'Status_Code' (0/1) is the column we want to predict.
y = df['Status_Code']
X = df.drop('Status_Code', axis=1)

# 2. Scaling Numerical Features
# Identify columns that need scaling (all features except the new binary Transaction_Type columns)
# We can simply select all remaining columns in X for safety, as StandardScaler works fine on 0s and 1s too.
numerical_cols = X.columns
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

print("✅ Features scaled successfully.")

# 3. Split the Data (Training and Testing)
# Use a 70/30 split for training and testing the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Data split: Training samples={len(X_train)}, Testing samples={len(X_test)}")

# 4. Train a Model (Using Random Forest Classifier)
# Random Forest is a robust choice for imbalanced fraud data
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

print("\n✅ Random Forest Model trained successfully.")
```

```python
# --- PHASE 3: MODEL EVALUATION ---

# 1. Make predictions on the test set
y_pred = model.predict(X_test)

# 2. Print the Classification Report and Confusion Matrix
print("\n--- Model Performance Evaluation ---")
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
# Note: The Confusion Matrix shows True Negatives, False Positives, False Negatives, True Positives
print(confusion_matrix(y_test, y_pred))
```

```python
import numpy as np 

# Assuming 'model' (Random Forest) and 'X' (Scaled Features) are still available

# 1. Extract Feature Importance
feature_importances = pd.Series(model.feature_importances_, index=X.columns)

# 2. Sort and select the Top 10 features
top_10_features = feature_importances.sort_values(ascending=False).head(10)

print("\n--- Top 10 Most Important Features ---")
print(top_10_features.to_string(float_format='{:.4f}'.format))
```
